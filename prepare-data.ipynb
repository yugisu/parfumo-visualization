```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/parfumo_data.csv')

df
```

```python
# Filter out perfumes wo name or brand.
df = df.dropna(subset=['Name', 'Brand'])

# Filter out perfumes that don't have any info about notes.
note_columns = ['Main_Accords', 'Base_Notes', 'Middle_Notes', 'Top_Notes']
df = df.dropna(subset=note_columns, how='all')
df = df.dropna(subset=['Main_Accords'], how='all')

df
```

```python
# Save the processed csv.
df.to_csv('src/assets/processed_parfumo_data.csv', index=False)
```

```python
# Extract unique note names.

perfume_notes = pd.concat([df[col] for col in note_columns]).str.split(',').explode().str.strip().unique()
perfume_notes = perfume_notes[~pd.isna(perfume_notes)]

"Floral" in perfume_notes
```

```python
import json

# with open('data/notes_to_color.json', 'r') as f:
#     notes_to_color = json.load(f)

# with open('data/notes_to_category.json', 'r') as f:
#     notes_to_category = json.load(f)

# notes_to_data = {k.lower(): {'color': c, 'category': notes_to_category[k]} for (k, c) in notes_to_color.items()}

with open('data/notes_to_note_details.json', 'r') as f:
    notes_to_data = json.load(f)

with open('data/perfume_notes_categories.json', 'r') as f:
    categories_data = json.load(f)
```

```python
from difflib import get_close_matches


def consolidate_similar_notes(notes_list, cutoff=0.9):
    recognized = notes_to_data.copy()
    missed = set()

    for note in notes_list:
        note = note.lower()

        if note in recognized or note in missed:
            pass

        # Find similar note
        similar = get_close_matches(note.lower(), recognized.keys(), n=1, cutoff=cutoff)

        if similar:
            recognized[note] = recognized[similar[0]]
        else:
            missed.add(note)

    return recognized, sorted(list(missed))


recognized, missed = consolidate_similar_notes(perfume_notes, 0.6)

if missed:
    raise RuntimeError("The missed list includes some notes!")
```

```python
recognized_flattened = {
    note: {'category': data["category"]}
    if categories_data[data["category"]]["color"] == data["color"]
    else data
    for (note, data) in recognized.items()
}

recognized_flattened
```

```python

with open('data/processed_notes_to_note_details.json', 'w') as f:
    json.dump(recognized_flattened, f)
```

```python
with open('missed.json', 'w') as f:
  json.dump({k: {'color': '', 'category': ''} for k in missed}, f)
```
